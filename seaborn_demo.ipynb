```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
sns.get_dataset_names()

tips=sns.load_dataset("tips")
# print(tips.head())
# print(tips.tail())

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    size='size',
    style='smoker')
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    kind='line',
    y='tip',
    hue='smoker',
    )
```
